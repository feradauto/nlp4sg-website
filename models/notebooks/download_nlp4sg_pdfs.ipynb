```python
from src.datamodules.apis.semantic_scholar_api import SemanticScholarAPI

def semantic_scholar_ids_to_arxiv_ids(ids):
    api = SemanticScholarAPI() 
    query = api.papers(ids, fields=['title', 'externalIds'])
    arxiv_ids = []
    for paper in query:
        if 'externalIds' in paper and 'ArXiv' in paper['externalIds']:
            arxiv_ids += [paper['externalIds']['ArXiv']]
        else:
            arxiv_ids += [None]
    return arxiv_ids 

nlp4sg_arxiv_ids = semantic_scholar_ids_to_arxiv_ids(nlp4sg_ids)
```

```python
import arxiv
nlp4sg_arxiv_filtered_ids = [aid for aid in nlp4sg_arxiv_ids if aid is not None]

def arxiv_ids_to_pdf_urls(ids):
    search = arxiv.Search(id_list = ids)
    pdf_urls = []
    for res in search.results():
        pdf_urls += [res.pdf_url]
    return pdf_urls
    
nlp4sg_arxiv_pdf_urls = arxiv_ids_to_pdf_urls(nlp4sg_arxiv_filtered_ids)
print(nlp4sg_arxiv_pdf_urls)
# ---- Download PDF
#import requests
#r = requests.get(pdf_url, allow_redirects=True)
#open(pdfdir, 'wb').write(r.content)
```

```python
import requests
from requests_futures.sessions import FuturesSession
from concurrent.futures import as_completed

def requests(urls):
    session = FuturesSession()
    futures=[session.get(url) for url in urls]
    return [future.result().content for future in as_completed(futures)]

response = requests(nlp4sg_arxiv_pdf_urls)
```
